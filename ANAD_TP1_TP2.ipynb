6) Créer un tableau noté
X(j) qui nous permet d'avoir pour chaque variable les informations suivantes : la moyenne, la variance, l'écart-type.

Nous pouvons utiliser pandas pour calculer ces statistiques.

```python
# Calcul des statistiques pour chaque variable
X_df = pd.DataFrame(X, columns=["H. Ball", "B. Ball", "Tennis", "Gym", "Natation", "F. Ball"])
#X_stats = X_df.describe().T[['mean', 'std', 'var']]

m = len(X_df)   # nombre d'individus

# 1 Calcul manuel des moyennes
moyennes = {}
for col in X_df.columns:
    somme = 0
    for valeur in X_df[col]:
        somme += valeur
    moyennes[col] = somme / m

# 2 Calcul manuel des variances (1/m)
variances = {}
for col in X_df.columns:
    somme_ecarts = 0
    for valeur in X_df[col]:
        somme_ecarts += (valeur - moyennes[col])**2
    variances[col] = somme_ecarts / m

# 3 Calcul manuel des écarts-types
ecarts_types = {}
for col in X_df.columns:
    ecarts_types[col] = np.sqrt(variances[col])

# 4 Création du tableau final
X_stats = pd.DataFrame({
    "Moyenne": moyennes,
    "Variance (1/m)": variances,
    "Ecart-type": ecarts_types
})

pd.set_option('display.float_format', '{:.4f}'.format)

print("Tableau X(j) avec formule du cours (1/m) :")
print(X_stats)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import linalg as LA
```

```python
# calcule de centre
def calcule_Centre(m):

  g = np.zeros(len(m[0]))
  for i in range(len(g)):
    g[i] = np.mean(m[:, i])
  return g
vect_g = calcule_Centre(X)
print('le centre de gravite = ', vect_g)
```

```python
#matrice centre_reduite
#d'abord on commence par calcule la matrice centree:
def matrice_centree(m):
  vec_g = calcule_Centre(m)
  c = np.zeros(m.shape)
  for i in range(len(m)):
    for j in range(len(m[i])):
      c[i][j] = m[i][j] - vec_g[j]
  return c

Y = matrice_centree(X)
print('matrice centre Y = \n', Y)
```

```python
def ecart_type(m):
  ecarttype = np.zeros(len(m[0]))

  for i in range(len(ecarttype)):

    ecarttype[i] = np.std(m[: , i])
  return ecarttype

ecartype = ecart_type(X)
ecartype
```

```python
def matrice_centree_reduite(m):
  mcr = np.zeros(m.shape)
  etp = ecart_type(X)
  for i in range (len(m)):
    for j in range(len(m[i])):
      mcr[i][j] = m[i][j] / etp[j]
  return mcr
Z = matrice_centree_reduite(Y)
print("matrice centre reduite Z = \n ", Z)
```

```python
# matrice de coorelation
def calcule_correlation(m):
  return ((1/len(m)) * (np.matmul(np.transpose(m), m)))

R = calcule_correlation(Z)
print('matrice de correlation = \n', R)
```

```python
corr = np.triu(R)
corr
```

```python
#La police utilisée
font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
        }
#la taille de la figure
plt.figure(figsize=(15, 7))
#le titre
plt.title("Matrice de correlation",fontdict=font)
#La representation
sns.heatmap(R,vmax=1,vmin=-1,annot = True, cmap = 'coolwarm',mask = corr)
#Le titre des axes
plt.xlabel('Xi', fontdict=font)
plt.ylabel('Xi', fontdict=font)
plt.show()
```

```python
# determination des valeurs propres et les vecteurs propres
valp, vecp = LA.eig(R)
print('les valeurs propres :\n', valp)
```

```python

print('les vecteurs propres :\n', vecp.T)
```

```python
# representation graphique des valeurs propres
plt.plot(sorted(valp, reverse=True))
plt.scatter(1,1.4, color = "r")
```

```python
# Trier les valeurs propres par ordre décroissant
valp_sorted = np.sort(valp)[::-1]

# Calcul de la variance expliquée
variance_expliquee = valp_sorted / np.sum(valp_sorted)

# Variance cumulée
variance_cumulee = np.cumsum(variance_expliquee)

print("Valeurs propres triées :\n", valp_sorted)
print("\nVariance expliquée (%) :\n", variance_expliquee * 100)
print("\nVariance cumulée (%) :\n", variance_cumulee * 100)

# Détermination du nombre d’axes pour atteindre 80%
nb_axes = np.argmax(variance_cumulee >= 0.80) + 1

print("\nNombre d'axes à retenir (>= 80%) :", nb_axes)

#lambda 1 + lambda 2 / somme des lambda = 81.7 >80 donc 2 axe sont suffisant
```

V5 et V20 se positionnent au centre-gauche avec un profil équilibré tous sports confondus, V5 affichant le niveau d'activité globale le plus élevé de l'échantillon. À l'opposé, V1 et V18 occupent la partie haute du graphique, traduisant une pratique dominée par le Tennis et le Football avec une activité générale modérée. V11 se singularise en bas du plan avec un profil Handball/Natation très marqué malgré une activité globale moyenne. Enfin, V15 apparaît comme la ville la moins sportive de l'échantillon, positionnée à droite-bas, loin de toutes les flèches de variables.
